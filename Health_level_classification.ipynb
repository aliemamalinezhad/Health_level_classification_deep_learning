# Health level classification of students 

###  Nursery UCI Dataset is used for classification

If you download the dataset it doesn't have any column name and it is very messy so you should first clean the dataset as below:

At first we load the row data which we just downlaoded. 

```python
import pandas as pd #we use pandas for data manipulation.

df = pd.read_csv('rawdata.csv')
df.head()
```

Now i am going to delete column with the name "1"

```python
df.drop("1", axis = 1, inplace = True)
```

```python
df.head()
```

Ok, now i set column to our data as follow:
I create a set of column and set it as column name:

```python
column_list= ["parents","has_nurs","form","children","housing","finance","social","health"]
df.columns =column_list
```

```python
df.head()
```

now we have to change the data types for example parents column are string we dont want strings in our data set so we change it to one hot dataset as follow:

```python
df.parents.value_counts()
```

```python
df_target = df.health
```

now we set out target we can remove it and create one-hot from other features to pass them for train process

```python
df.drop("health", axis = 1, inplace = True)
```

```python
df.head()
```

As you see now we dont have health column now we make one-hot from other features

```python
df_one_hot = pd.get_dummies(df)
```

```python
df_one_hot.head()
```

OK, now we have a clean one hot matrix with "non" data we chack it again as follow:

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

As you see all features are without non data so its great.

Now we try to concat data to prepare our dataset.

```python
dataset = pd.concat([df_target,df_one_hot],axis = 1)
```

```python
dataset.head()
```

At last we just categorize targets .

```python
dataset.replace("not_recom",0 ,inplace = True)
dataset.replace("priority",1 ,inplace = True)
dataset.replace("spec_prior",2 ,inplace = True)
dataset.replace("very_recom",3 ,inplace = True)
dataset.replace("recommend",4 ,inplace = True)
```

```python
dataset.head()
```

```python
clean_dataste = dataset.to_csv("clean_data.csv")
```

```python
#def load_data(input_file):
def train_evaluate(model, x_train, y_train, x_test, y_test):
    model.fit(x_train, y_train,batch_size=32, nb_epoch=1000)
    return model.evaluate(x_test, y_test)

df =pd.read_csv('clean_data.csv')

X = np.array(df.drop(['health'], 1)) 
df.dropna(inplace=True)
y = np.array(df['health'])


Y = to_categorical(y)
print(Y)
X_train, X_test, y_train, y_test =train_test_split (X, Y, test_size=0.2)


print("categorized target = ",Y)

print('y_test = ',y_test)

max_features = 1000000
maxlen = 202
embedding_size = 320






batch_size = 100


epochs = 1000


def create_model():
  # def create_model(input_length):
    print ('Creating model...')
    model = Sequential()
    model.add(BatchNormalization())
    model.add(Dense(512, input_dim=34, init='uniform'))
    model.add(Dropout(0.5))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.5))
    model.add(Dense(256, init='uniform'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.5))
    model.add(Dense(5, init='uniform'))
    model.add(Activation('softmax'))

    #sgd = SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
    #adam = Adam(lr=0.05)


    print ('Compiling...')
    
    
    model.compile(loss='categorical_crossentropy',
                  optimizer=Adam(lr=0.01, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False),
                  #metrics=['accuracy'])
                  metrics=['accuracy'])



  
    return model





print(X_train.shape)

model = create_model()

print ('Fitting model...')
history = model.fit(X_train, y_train, batch_size=1000, nb_epoch=100, validation_split = 0.1, verbose = 1)


# plt.show()
score, acc = model.evaluate(X_test, y_test, batch_size=100)
y_pred = model.predict(X_test)
print('acc = ', acc)
print(y_pred)
y_pred = (y_pred > 0.5) 


print(classification_report(y_test, y_pred))

print(y_pred)
model.summary()

```
